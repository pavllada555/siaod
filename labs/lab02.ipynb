```python
def binary_search(array, element):
    mid = 0
    start = 0
    end = len(array)
    while(start<= end):
        mid = (start + end) // 2

        if element == array[mid]:
            return mid
        if element < array[mid]:
            end = mid - 1
        else:
            start = mid + 1
    return None
```

```python
class BSTNode:
    def __init__(self, value: float | int = None, index: int | set[int] = None):
        self.left = None
        self.right = None
        if value is None or index is None: raise TypeError
        self.value: float | int = value
        self.index: set[int]
        if type(index) is set:
            self.index = index
        elif type(index) is int:
            self.index = {index}

    def insert(self, other):
        if self == other:
            self.index.update(other.index)
            return
        if other < self:
            if self.left is not None:
                self.left.insert(other)
                return
            self.left = other
        else:
            if self.right is not None:
                self.right.insert(other)
                return
            self.right = other
    

    def search(self, other):
        if self == other:
            return self
        if self.right is not None and other > self:
            return self.right.search(other)
        if self.left is not None and other < self:
            return self.left.search(other)
        return None
        

    def __eq__(self, other):
        if type(other) is BSTNode:
            return self.value == other.value
        return False

    def __lt__(self, other):
        if type(other) is BSTNode:
            return self.value < other.value
        return False

    def __gt__(self, other):
        if type(other) is BSTNode:
            return self.value > other.value
        return False

        
    def delete(self, val):
        if self == None:
            return self
        if val < self.val:
            if self.left:
                self.left = self.left.delete(val)
            return self
        if val > self.val:
            if self.right:
                self.right = self.right.delete(val)
            return self
        if self.right == None:
            return self.left
        if self.left == None:
            return self.right
        min_larger_node = self.right
        while min_larger_node.left:
            min_larger_node = min_larger_node.left
        self.val = min_larger_node.val
        self.right = self.right.delete(min_larger_node.val)
        return self
```

```python
class BST:
    def __init__(self, massiv: list[int | float] = None):
        self.massiv = sorted(massiv) if massiv is not None else []
        self.middle = len(self.massiv) // 2
        self.tree = BSTNode(self.massiv[self.middle], self.middle)
        for index, element in enumerate(self.massiv):
            self.tree.insert(BSTNode(element, index))

    def search(self, value: int | float):
        search_result = self.tree.search(BSTNode(value, 0))
        if search_result is not None:
            return search_result.index
        return None
binary_tree = BST(array)
```

```python
def fibonacchiSearch(array, element): #arr = array, element = number, n = lenght of array
    n = len(array)
    fibMMm2 = 0 #(m-2)ое число Фибоначчи
    fibMMm1 = 1 #(m-1)ое число фибоначчи
    fibM = fibMMm2 + fibMMm1 # мное число фибоначчи
    while (fibM < n): 
        fibMMm2, fibMMm1 = fibMMm1, fibM
        fibM = fibMMm2 + fibMMm1
    offset = -1
    while (fibM>1):
        i = min(offset+fibMMm2, n-1)
        if(array[i]<element):
            fibM, fibMMm1  = fibMMm1, fibMMm2
            fibMMm2 = fibM - fibMMm2
            offset = i
        elif(array[i]>element):
            fibM = fibMMm2
            fibMMm1 = fibMMm1 - fibMMm2
            fibMMm2 = fibM - fibMMm1
        else:
            return i
    if(fibMMm1 and array[n-1]==element):
        return n-1
    return None
```

```python
def IntrepolationSearch(array, element):
    low = 0
    high = len(array) - 1
    while low <= high and element >= array[low] and element <= array[high]:
        index =  low + int(((float (high-low)/(array[high]-array[low])) * (element-array[low])))
        if array[index] == element:
            return index
        if array[index] < element:
            low = index + 1
        else:
            high = index - 1
    return None
```

```python
def NumPySearch(array, element):
    return np.where(array==element)
```

```python
def Standart(array, element):
    return array.index(element, 0, len(array))
```

```python

# Класс хеш-таблицы с методом простого рехеширования
class HashTableEasy:
    # Конструктор класса (инициализация значений)
    def __init__(self, table_size):
        self.table_size = table_size#размер таблицы
        self.table = np.zeros((self.table_size, 2))#заполняем нулями
    
    # Функция хеширования
    def __hash__(self, value):
        result = 1
        while (value > 0):
            m = value % 10
            result = (result * m) % self.table_size
            value = value // 10
        return result#получам хэш-значение
    
    # Метод добавления элемента в хеш-таблицу
    def add(self, value):
        h1 = self.__hash__(value)
        if (self.table[h1][1] == 0):#если клетка свободна, то
            self.table[h1][0] = value#заполняем значением
            self.table[h1][1] = 1#и помечаем, что клетка занята
        else:
            i = 0#счетчик
            h2 = h1#чтобы идти не с нуля, а с последнего занятого
            while (self.table[h2][1] != 0 and i <= self.table_size):#пока не освободится или не выйдет за пределы таблицы
                i = i + 1#прибавляем счетчик
                h2 = (h1 + i) % self.table_size#получаем ключ от ячейки
            if (i > self.table_size): print("Map is full")#если вышли из пределов, то пишем, что таблица заполнена
            else:
                self.table[h2][0] = value#иначе присваиваем значения и
                self.table[h2][1] = 1#занимаем ячейку
    
    # Метод поиска элемента в хеш-таблице
    def find(self, value):
        h1 = self.__hash__(value)#получаем хэш-значение
        if (self.table[h1][1] == 1):#и проверяем, занята ли ячейка по данному хэшу
            if (self.table[h1][0] == value): return h1#если искомое и найденное совпадают, то возвращаем ключ
            else:#иначе ищем нужное нам значение в других хэшключах
                i = 0
                h2 = h1
                while (self.table[h2][1] == 1 and i < self.table_size):
                    i = i + 1
                    h2 = (h1 + i) % self.table_size
                    if (self.table[h2][0] == value): return h2
                    if (self.table[h2][1] == 0): return -1
        else:
            return -1
    
    # Метод удаления элемента в хеш-таблице
    def remove(self, value):
        index = self.find(value)#находим индекс нужного значения
        if (index != 1):
            self.table[index][1] = 0#помечаем, что ячейка освободилась
            print("Element is deleted")
        else: print("Element not found")#иначе, что элемент не найден
    
    # Метод для вывода таблицы
    def show(self):
        print(self.table)#простой принт
    
    # Метод для очищения таблицы
    def clear(self):
        self.table = np.zeros((self.table_size, 2))#заполняем таблицу ноликами
```

```python
# Класс хеш-таблицы с рехешированием с помощью псевдослучайных чисел
class HashTableRand:
    # Конструктор класса (инициализация значений)
    def __init__(self, table_size):
        self.table_size = table_size#размер
        self.table = np.zeros((self.table_size, 2))#заполняем нулями
        # Массив псевдослучайных чисел
        self.rand = np.zeros(self.table_size)
        self.rand = self.rand.astype(int)
        for i in range(1, self.table_size):
            self.rand[i] = int(random.randrange(1, 1000))
    
    # Функция хеширования
    def __hash__(self, value):
        result = 1
        while (value > 0):
            m = value % 10#деление по модулю
            result = (result * m) % self.table_size
            value = value // 10#целочисленное
        return result#получаем значение хэша
    
    # Метод добавления элемента в хеш-таблицу
    def add(self, value):
        h1 = self.__hash__(value)#берем хэшключ
        if (self.table[h1][1] == 1):#если занята ячейка, то
            i = 0#счетчик
            while (self.table[h1][1] == 1 and i < self.table_size):
                h1 = (h1 + self.rand[i]) % self.table_size#получаем другой хэшключ, благодаря псевдослучайным числам
                i = i + 1#и прибавляем единицу к счетчику
            if (i >= self.table_size):#если счетчик больше или равен размера таблицы
                # Линейный поиск
                h1 = 0
                while (self.table[h1][1] != 0 and h1 < self.table_size):#проходимся по таблице и смотрим, есть ли пустые ячейки
                    h1 = h1 + 1#и меняем хэшключ
                if (h1 < self.table_size):#если ячейка найдена
                    self.table[h1][0] = value#заполняем
                    self.table[h1][1] = 1
                else: print("Map is full")#иначе пишем, что таблица заполнена
            else:
                self.table[h1][0] = value#если же счетчик не превышает размер таблицы, то
                self.table[h1][1] = 1#заполняем ту ячейку, которую нашли ранее
        else:
            self.table[h1][0] = value#если же ИЗНАЧАЛЬНО ячейка была пустая, то заполняем
            self.table[h1][1] = 1#в самую-самую первую ячейку
    
    # Метод поиска элемента в хеш-таблице
    def find(self, value):#поиск по хэштаблице как и раньше
        h1 = self.__hash__(value)
        if (self.table[h1][1] == 1):
            if (self.table[h1][0] == value): return h1
            else:
                i = 0
                while (self.table[h1][1] == 1 and i < self.table_size):
                    h1 = (h1 + self.rand[i]) % self.table_size
                    i = i + 1
                    if (self.table[h1][0] == value): return h1
                    if (self.table[h1][1] == 0): return -1
        else:
            return -1
    
    # Метод удаления элемента в хеш-таблице
    def remove(self, value):#удаление как и раньше
        index = self.find(value)
        if (index != 1):
            self.table[index][1] = 0
            print("Element is deleted")
        else: print("Element not found")
    
    # Метод для вывода таблицы
    def show(self):#простой принт
        print(self.table)
    
    # Метод для очищения таблицы
    def clear(self):#простое очищение
        self.table = np.zeros((self.table_size, 2))
```

```python
# Класс хеш-таблицы с рехешированием методом цепочки (БУДЕТ СПИСОК С МАССИВАМИ ЗНАЧЕНИЙ)
class HashTableСhaining:
    # Конструктор класса (инициализация значений)
    def __init__(self, table_size):
        self.table_size = table_size#размер
        self.table = []#делаем ее пустым списком
        for i in range(0, self.table_size):
            self.table.append(None)#заполняем нонами
    
    # Функция хеширования
    def __hash__(self, value):
        result = 1
        while (value > 0):
            m = value % 10#по модулю
            result = (result * m) % self.table_size
            value = value // 10#целочисленное
        return result#получаем хэшключ
    
    # Метод добавления элемента в хеш-таблицу
    def add(self, value):
        h1 = self.__hash__(value)#получаем хэш
        if (self.table[h1] is None):#если ячейка ноновская
            self.table[h1] = [value]#просто меняем значение
        else:
            i = self.table[h1].append(value)#иначе добавляем элемент как элемент массива
    
    # Метод поиска элемента в хеш-таблице
    def find(self, value):
        h1 = self.__hash__(value)#получаем хэш
        if (self.table[h1] is None):#если по хэшу none то возвращаем ошибку
            return -1
        else:
            if (self.table[h1].count(value) > 0): return h1#иначе смотрим, сколько элементов в массиве по хэшключу
            # и если больше нуля, то возвращаем ключ
            else: return -1#иначе - ошибка
    
    # Метод удаления элемента в хеш-таблице
    def remove(self, value):
        index = self.find(value)#ищем индекс
        if (index != 1):
            self.table[index].remove(value)#убираем значение по индексу
            print("Element is deleted")
        else: print("Element is not found")
    
    # Метод для вывода таблицы
    def show(self):
        print(self.table)
    
    # Метод для очищения таблицы
    def clear(self):
        self.table = np.zeros((self.table_size, 2))
```
